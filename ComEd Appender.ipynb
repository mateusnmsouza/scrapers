## Python 3 code to append resulting files from ComEd electricity usage data

The following code cell is the preamble to load required packages:

```python
import os
import csv
import pandas as pd
import numpy as np
import zipfile
```

Define the file paths where data will be saved, or loaded from:

```python
unzipped_path = 'C:/Users/Mateus/OneDrive/Documentos/AnacondaProjects/ComEdScraper_020818/Results/newunzip/'    # where UNZIPPED results of non-faulty accounts are saved
zipped_path = 'C:/Users/Mateus/OneDrive/Documentos/AnacondaProjects/ComEdScraper_020818/Results/'    # where results of non-faulty accounts are saved
master_path = 'C:/Users/Mateus/OneDrive/Documentos/AnacondaProjects/ComEdScraper_020818/ComEd_Data_Feb2018.csv'    # where the master appended file will be saved
```

Unzip all the files:

```python
filecount = 0
for filename in os.listdir(zipped_path):
    filecount+=1
    print("File number {} is {}".format(filecount, filename))
        
    comed_zip = zipfile.ZipFile(zipped_path+filename)
    comed_zip.extractall(unzipped_path)
    comed_zip.close()
```

Code to append all files obtained from the scraper, storing results into the specified filepath:

```python
masterCSV = open(master_path, 'a')
masterWriter = csv.writer(masterCSV, delimiter=",", lineterminator='\n')
masterWriter.writerow(["account_number", "rate", "endOfBilling", "daysInBilling", "totalKWH_Usage", "onPeakKWH_Usage", "offPeakKWH_Usage", "billingDemandKW", "monthlyPeakDemandKW"])

filecount = 0
for filename in os.listdir(unzipped_path):
    filecount+=1
    if filename.endswith(".csv"):
        oldCSV = open(unzipped_path+filename)
        rows = csv.reader(oldCSV)
        rowNum = 0
        accountNum = 0
        masterRow = 1
        for row in rows:
            if rowNum == 0:
                # accountNum.append(int(row[1]))
                accountNum = int(row[1])
            if rowNum > 24 and rowNum < 49:
                newRow = []
                newRow.append(accountNum)
                rowIndex = 0
                valid = True
                while(rowIndex < 8):
                    try:
                        newRow.append(row[rowIndex])
                        rowIndex += 1
                    except IndexError:
                        rowIndex +=1
                        valid = False
                        continue
                if(valid):
                     masterWriter.writerow(newRow)
            rowNum += 1 
        oldCSV.close()
        print("File number {} is account number {}".format(filecount, accountNum));
        continue
    else:
        continue
masterCSV.close()
```
